# Train-valid-test split - Full data

```python
RANDOM_STATE = 42
VALID_SIZE = 0.2
TEST_SIZE = 0.2

DATA_DIR = '../../data/raw/full/'
DATA_PATH = DATA_DIR + 'dataset_full.csv'
```

## Load data

```python
import pandas as pd

all_data = pd.read_csv(DATA_PATH)

all_data.info()
all_data.head()
```

## Split

```python
from sklearn.model_selection import train_test_split

TARGET_COL_NAME = 'phishing'

train_data, valid_test_data = train_test_split(all_data, test_size=VALID_SIZE+TEST_SIZE, stratify=all_data[TARGET_COL_NAME], random_state=RANDOM_STATE, shuffle=True)
```

```python
train_data.info()
train_data.head()
```

```python
valid_test_data.info()
valid_test_data.head()
```

```python
valid_data, test_data = train_test_split(valid_test_data, test_size=0.5, stratify=valid_test_data[TARGET_COL_NAME], random_state=RANDOM_STATE, shuffle=True)
```

```python
valid_data.info()
valid_data.head()
```

```python
test_data.info()
test_data.head()
```

```python
print('All data: {}'.format(all_data.shape))
print('Training data: {}'.format(train_data.shape))
print('Validation data: {}'.format(valid_data.shape))
print('Test data: {}'.format(test_data.shape))
```

```python
print('All data class ratio:')
all_data_target_cnts = all_data[TARGET_COL_NAME].value_counts()
print('Phishing/non-phishing: {:.3f}'.format(all_data_target_cnts[1] / all_data_target_cnts[0]))

all_data_target_cnts
```

```python
print('Training data class ratio:')
train_data_target_cnts = train_data[TARGET_COL_NAME].value_counts()
print('Phishing/non-phishing: {:.3f}'.format(train_data_target_cnts[1] / train_data_target_cnts[0]))

train_data_target_cnts
```

```python
print('Validation data class ratio:')
valid_data_target_cnts = valid_data[TARGET_COL_NAME].value_counts()
print('Phishing/non-phishing: {:.3f}'.format(valid_data_target_cnts[1] / valid_data_target_cnts[0]))

valid_data_target_cnts
```

```python
print('Test data class ratio:')
test_data_target_cnts = test_data[TARGET_COL_NAME].value_counts()
print('Phishing/non-phishing: {:.3f}'.format(test_data_target_cnts[1] / test_data_target_cnts[0]))

test_data_target_cnts
```

## Save split data

```python
train_data.to_csv(DATA_DIR + 'train.csv')
valid_data.to_csv(DATA_DIR + 'valid.csv')
test_data.to_csv(DATA_DIR + 'test.csv')
```
