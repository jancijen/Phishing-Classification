# PCA - Small

```python
import pandas as pd
```

## Load Data

```python
DATA_DIR = '../../data/'
DATA_SIZE_DIR = '../../data/raw/small/'
```

```python
import pandas as pd

train_data = pd.read_csv(DATA_SIZE_DIR + 'train.csv')

train_data.info()
train_data.head()
```

```python
valid_data = pd.read_csv(DATA_SIZE_DIR + 'valid.csv')

valid_data.info()
valid_data.head()
```

```python
test_data = pd.read_csv(DATA_SIZE_DIR + 'test.csv')

test_data.info()
test_data.head()
```

## Get feature data

```python
TARGET_COL_NAME = 'phishing'
```

```python
X_train = train_data.drop(TARGET_COL_NAME, axis=1)
X_valid = valid_data.drop(TARGET_COL_NAME, axis=1)
X_test = test_data.drop(TARGET_COL_NAME, axis=1)
```

## Dimensionality reduction - PCA

```python
from sklearn.decomposition import PCA

num_components = 0.99

pca = PCA(n_components=num_components, svd_solver='full')
```

```python
X_train_pca = pd.DataFrame(pca.fit_transform(X_train))

X_train_pca.info()
X_train_pca.head()
```

```python
X_valid_pca = pd.DataFrame(pca.transform(X_valid))

X_valid_pca.info()
X_valid_pca.head()
```

```python
X_test_pca = pd.DataFrame(pca.transform(X_test))

X_test_pca.info()
X_test_pca.head()
```

## Save reduced datasets

```python
pd.concat([X_train_pca, train_data[TARGET_COL_NAME]], axis=1).to_csv(DATA_DIR + 'reduced/small/PCA_train.csv', index=False)
pd.concat([X_valid_pca, valid_data[TARGET_COL_NAME]], axis=1).to_csv(DATA_DIR + 'reduced/small/PCA_valid.csv', index=False)
pd.concat([X_test_pca, test_data[TARGET_COL_NAME]], axis=1).to_csv(DATA_DIR + 'reduced/small/PCA_test.csv', index=False)
```
